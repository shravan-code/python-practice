# zoneinfo

The zoneinfo module, introduced in Python 3.9, provides first-class support for IANA time zones (e.g., Asia/Kolkata, UTC, America/New_York) without relying on third-party libraries like pytz.

It is the recommended standard for timezone-aware datetime handling in modern Python.

```python
from zoneinfo import ZoneInfo, available_timezones
from datetime import datetime
import zoneinfo
```

```python
print(zoneinfo.__all__)
```

```python
datetime.now(tz=ZoneInfo("Asia/Kolkata"))
```

```python
print("Asia/Kolkata" in available_timezones())
```

```python
available_timezones()
```

```python
from datetime import datetime
from zoneinfo import ZoneInfo

UTC = ZoneInfo("UTC")
IST = ZoneInfo("Asia/Kolkata")

now_utc = datetime.now(tz=UTC)
now_ist = now_utc.astimezone(IST)
now_ist
```

```python
from datetime import datetime, timezone, timedelta

IST = timezone(timedelta(hours=5, minutes=30))
datetime.now(tz=IST)
```
