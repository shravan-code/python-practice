# builtins module

The builtins module defines the core namespace of Python. Every Python program implicitly has access to these names without any import statements. In practical terms, builtins contains the languageâ€™s most fundamental functions, types, exceptions, and constants.

```python
import builtins
```

```python
builtins.len([1, 2, 3, 4, 5]), len([1, 2, 3, 4, 5])
```

```python
builtins.print("shravan"), print("shravan")
```

```python
sum([1, 2, 3, 4, 5]), builtins.sum([1, 2, 3, 4, 5])
```

```python
print(dir(builtins))
```

```python
builtins.display("Hello, World!")
```

```python
isinstance(42, builtins.int), builtins.isinstance(42, int)
```
