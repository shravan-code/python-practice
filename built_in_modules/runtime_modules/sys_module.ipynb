# sys module

The sys module provides direct access to interpreter-level variables and functions. It allows you to interact with the Python runtime environment, command-line arguments, memory management, module loading, and program termination.

**In short:**

sys = control + introspection of the Python interpreter

```python
import sys
```

```python
sys.argv[0]
```

```python
sys.exit(1)
```

```python
sys.version
```

```python
sys.version_info
```

```python
sys.api_version
```

```python
sys.getwindowsversion()
```

```python
sys.executable
```

```python
sys.path.append('/my/custom/path')
```

```python
print("error message", file=sys.stdout)
```

```python
x = [1, 2, 3, 4, 5]

sys.getsizeof(x)
```

```python
sys.getrecursionlimit()
```

```python
sys.getrefcount(x), sys.getdefaultencoding(), sys.getfilesystemencoding()
```

```python
sys.modules.keys()
```

```python
sys.platform, sys.byteorder, sys.maxsize
```

| Module       | Purpose                    |
| ------------ | -------------------------- |
| `sys`        | Interpreter internals      |
| `os`         | OS-level operations        |
| `argparse`   | CLI argument parsing       |
| `platform`   | OS & hardware info         |
| `subprocess` | External command execution |
