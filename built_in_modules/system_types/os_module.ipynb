# os module

The os module provides a portable interface to operating systemdependent functionality.
 It allows Python programs to interact with the underlying OS for file systems, processes, environment variables, permissions, and paths, while abstracting platform differences (Linux, macOS, Windows).


### File & Directory Operations

Used to inspect, create, remove, and traverse the filesystem.

* os.getcwd()  current working directory
* os.chdir(path)  change working directory
* os.listdir(path)  list directory contents
* os.mkdir(path), os.makedirs(path)  create directories
* os.remove(path)  delete a file
* os.rmdir(path), os.removedirs(path)  delete directories
* os.rename(src, dst)  rename files/directories
