# venv module

The venv module (introduced in Python 3.3) is the standard library solution for creating isolated Python environments. Each virtual environment contains its own Python interpreter, pip, and site-packages directory, independent of the system-wide Python installation.

**One-line definition:**
venv provides dependency isolation so each project can have its own Python packages and versions.

When you create a virtual environment:

* A copy or symlink of the Python interpreter is created
* A private site-packages directory is initialized
* pip is bootstrapped locally

Environment activation modifies:

        PATH
        VIRTUAL_ENV
        Python import resolution order

Python then resolves imports inside the venv first, not globally.
