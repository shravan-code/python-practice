# argparse module

The argparse module is Pythonâ€™s standard library for building robust command-line interfaces (CLIs). It parses command-line arguments, validates inputs, generates help/usage messages, and provides a clean API for CLI-driven programs.

* Declarative definition of CLI arguments
* Automatic --help and usage text
* Type conversion and validation
* Support for positional and optional arguments
* Subcommands (e.g., git commit, git push)
* Industry-standard for production-grade scripts and tools

```python
import argparse
import pathlib
```

```python
pathlib.Path("test.csv")
```

```python
parser = argparse.ArgumentParser(
    description="A simple CLI tool example using argparse")
parser
```

```python
parser.add_argument("input_file", help="Input csv file path")
```

```python
parser.add_argument(
    "--env",
    choices=["dev", "prod"],
    default="dev",
    help="Environment to run the script in (default: dev)"
)
```

```python
parser.add_argument('--batch-size', type=int, default=32,
                    help='Batch size for processing (default: 32)')
```
