The decimal module provides decimal floating-point arithmetic with user-defined precision and exact decimal representation. It is specifically designed for financial, accounting, and high-precision scientific calculations where binary floating-point (float) is unsuitable.

```python
import decimal
```

```python
print(dir(decimal))
```

```python
0.1 + 0.2
```

```python
decimal.Decimal('0.1') + decimal.Decimal('0.2')
```

```python
decimal.Decimal(0.1)
```

```python
decimal.Decimal('0.1')
```

```python
decimal.getcontext().prec = 10
decimal.Decimal('1') / decimal.Decimal('3')
```
