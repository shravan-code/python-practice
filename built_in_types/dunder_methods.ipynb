    __class__

**What**: The class an object is instantiated from

**Use**: Runtime type inspection

```python
x = 10
x.__class__  # <class 'int'>
```

```python
print(x.__class__)
```

    __dict__

**What**: Writable attribute namespace of an object

**Use**: Dynamic attributes, serialization

```python
class A:
    pass


a = A()
a.x = 5
a.__dict__  # {'x': 5}
```

    __doc__

**What**: Documentation string

**Use**: Help systems, introspection

```python
def f():
    """My function"""
    pass


f.__doc__  # 'My function'
```

    __module__

**What**: Module where object is defined

**Use**: Debugging, reflection

```python
import math

math.sqrt.__module__  # 'math'
```

    __repr__

**What**: Developer-facing representation

**Use**: Debugging

```python
class User:
    def __repr__(self):
        return "Hi this is shravan"


repr(User())  # Hi this is shravan
```

    __str__

**What**: Human-readable string

**Use**: Logging, UI output

```python
class User:
    def __str__(self):
        return "User Object"


print(User())  # User Object
```

    __hash__

**What**: Hash value for sets/dicts

**Use**: Hash-based collections

```python
hash("abc")
```

    __eq__, __ne__, __lt__, __le__, __gt__, __ge__

**What**: Comparison operators

**Use**: Sorting, equality checks

```python
class A:
    def __eq__(self, other):
        return True


A() == A()  # True
```

__name__

What: Class or function name

```python
class A:
    pass


A.__name__  # 'A'
```

__bases__

What: Parent classes
Use: Inheritance analysis

```python
class A: pass


class B(A): pass


B.__bases__  # (<class '__main__.A'>,)
```

__mro__

What: Method Resolution Order
Use: Multiple inheritance resolution

```python
B.__mro__
```

__subclasses__()

What: All subclasses of a class
Use: Plugin systems

```python
A.__subclasses__()
```

__defaults__

What: Default parameter values

```python
def f(a=10):
    pass


f.__defaults__  # (10,)
```

__kwdefaults__

What: Keyword-only defaults

```python
def f(*, x=5):
    pass


f.__kwdefaults__  # {'x': 5}
```

__annotations__

What: Type hints

```python
def f(a: int) -> str:
    pass


f.__annotations__
```

__code__

What: Compiled bytecode
Use: Advanced introspection

```python
def f(): pass


f.__code__.co_varnames
```

__closure__

What: Closed-over variables
Use: Decorators, closures

```python
def outer():
    x = 10

    def inner():
        return x

    return inner


outer().__closure__[0].cell_contents  # 10
```

__call__

What: Makes an object callable
Use: Function-like objects

```python
class A:
    def __call__(self):
        return "called"


A()()  # 'called'
```

__getattribute__

What: Intercepts all attribute access

```python
class A:
    def __getattribute__(self, name):
        return "intercepted"


A().x  # 'intercepted'
```

__getattr__

What: Fallback for missing attributes

```python
class A:
    def __getattr__(self, name):
        return "missing"


A().x  # 'missing'
```

__setattr__

What: Intercepts assignment

```python
class A:
    def __setattr__(self, name, value):
        print(name, value)


A().x = 10
```

__delattr__

What: Intercepts deletion

```python
class A:
    def __delattr__(self, name):
        print("deleted", name)
```

__len__

What: Length of object

```python
class A:
    def __len__(self):
        return 5


len(A())  # 5
```

__getitem__

What: Index access

```python
class A:
    def __getitem__(self, i):
        return i * 2


A()[3]  # 6
```

__setitem__, __delitem__

What: Assignment / deletion

```python
class A:
    def __setitem__(self, k, v):
        print(k, v)
```

__iter__

What: Returns iterator

```python
class A:
    def __iter__(self):
        return iter([1, 2, 3])
```

__next__

What: Iterator step

```python
class A:
    def __iter__(self):
        return self

    def __next__(self):
        raise StopIteration
```

__enter__

What: Setup block

__exit__

What: Cleanup block

```python
class A:
    def __enter__(self):
        print("start")

    def __exit__(self, exc_type, exc, tb):
        print("end")


with A():
    pass
```
