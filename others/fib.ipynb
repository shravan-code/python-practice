```python

def fib(n):
    a, b = 0, 1
    result = []
    for _ in range(n):
        result.append(a)
        a, b = b, a + b
    return result


fib(10)
```

```python
def fib(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b


list(fib(10))
```

```python
def fib_recursive(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib_recursive(n - 1) + fib_recursive(n - 2)


[fib_recursive(i) for i in range(10)]
```

```python
def fib_memoized(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        memo[n] = fib_memoized(n - 1, memo) + fib_memoized(n - 2, memo)
        return memo[n]


[fib_memoized(i) for i in range(10)]
```

```python
def fib_dynamic(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    fib_sequence = [0, 1]
    for i in range(2, n):
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
    return fib_sequence


fib_dynamic(10)
```

```python
def fibs(n):
    a = 0
    b = 1
    c = 0
    while c < n:
        yield a
        a, b = b, a + b
        c += 1


x = fibs(10)
list(x)
```
