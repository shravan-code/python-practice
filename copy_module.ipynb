```python
import copy
```

```python
a = [1, 2, [3, 4]]
b = a
```

```python
b.append(10)
```

```python
b, a
```

```python
id(a) , id(b)
```

b is not a new list. It points to the same list in memory.
Changes to b modify a.

To avoid this, we use shallow or deep copies.

**DeepCopy - independent copy**

```python
c = [1,2,3]
d = copy.deepcopy(c)
c.append(4)
d.append(5)
c,d
```

```python
id(c) , id(d)
```

**ShallowCopy - it copies and references nested objects only**

```python
e = [1,2,3]
f = copy.copy(e)
e.append(4)
f.append(5)
e,f,id(e),id(f)
```

```python
e = [1,2,[6,7]]
f = copy.copy(e)
e.append(4)
f.append(5)
e,f,id(e),id(f)
```
