**itertools Module (Python Standard Library)**

**Overview**

The itertools module provides a collection of fast, memory-efficient iterator-building blocks.
It is designed for high-performance looping, especially useful in:

* Large-scale data processing
* Streaming pipelines
* Combinatorics
* Functional-style programming
* Data engineering workloads (joins, windows, rolling logic)

**Key design principles:**
* Lazy evaluation (values generated on demand)
* Constant memory usage
* Composable iterators

**Count(start, step)**

```python
import itertools as tt

x = tt.count(10, 2)

for i in x:
    print(i)
    if i >= 20:
        break
```

```python
# do this without modules
def accumulate_mul(iterable):
    total = iterable[0]
    for item in iterable:
        total *= item
        yield total


list(accumulate_mul([1, 2, 3, 4]))
```

to fully flatten a list, Recursive method is best

```python
def flatten_list(nested_list):
    for item in nested_list:
        if isinstance(item, (list, tuple)):
            yield from flatten_list(item)
        else:
            yield item


nested_list = [1, [2, [3, 4]], 5, [6, [7, (10, [11, (12, 13)])]], 9]
list(flatten_list(nested_list))
```

**dropwhile(predicate, iterable)**

```python
x = tt.dropwhile(lambda n: n < 5, [1, 4, 6, 4, 1])
list(x)  # drops until first predicate is false i.e, 6 < 5 is false so from 6 onwards it yields all


# where as filter applies to all elements , dropwhile only until first false

def dropping(p, iterable):
    it = iter(iterable)
    for item in it:
        if not p(item):
            yield item
            break

    for item in it:
        yield item


list(dropping(lambda n: n < 5, [1, 4, 6, 4, 1]))
```

**takewhile(predicate, iterable)**

```python
list(tt.takewhile(lambda x: x < 5, [1, 8, 3, 5, 7]))


# code without modules
def taking(p, iterable):
    for item in iterable:
        if p(item):
            yield item
        else:
            break


list(taking(lambda n: n < 5, [1, 8, 3, 5, 7]))
```

**pairwise(iterable)**

```python

list(tt.pairwise([1, 2, 3, 4]))


# code without modules
def pairwise_no_module(iterable):
    it = iter(iterable)
    prev = next(it)
    for curr in it:
        yield (prev, curr)
        prev = curr


list(pairwise_no_module([1, 2, 3, 4]))

x = [1, 2, 3, 4]
list(zip(x, x[1:]))
```

**product(iterables, repeat=1)**

```python

list(tt.product([1, 2], ["a", "b"]))


# code without modules
def product_no_module(*iterables, repeat=1):
    pools = [list(pool) for pool in iterables] * repeat
    result = [[]]
    for pool in pools:
        result = [x + [y] for x in result for y in pool]
    for prod in result:
        yield tuple(prod)


list(product_no_module([1, 2], ["a", "b"]))
```

**permutations(iterable, r=None) - does not follow order - unordered numbers**

```python

list(tt.permutations([1, 2, 3], 3))
```

```python
x = [1, 2, 3]


def recursive_permutations(lst, r=None):
    if r is None:
        r = len(lst)
    if r == 0:
        yield ()
    else:
        for i in range(len(lst)):
            for p in recursive_permutations(lst[:i] + lst[i + 1:], r - 1):
                yield (lst[i],) + p


list(recursive_permutations(x, 2))
```

```python
def fib(n):
    a, b = 0, 1
    result = []
    for _ in range(n):
        result.append(a)
        a, b = b, a + b
    print(result)
    return a


fib(8)
```

```python
# palindrome
def is_palindrome(s):
    return s == s[::-1]


is_palindrome("radar")
```

**combinations(iterable, r) - follow order of the numbers**

```python
x = [1, 2, 3, 4]
list(tt.combinations(x, 2))  # (1,2),(1,3),(1,4),(2,3),(2,4),(3,4)

for i in x:
    for j in x:
        if i < j:
            print((i, j))
```

**combinations_with_replacement(iterable, r)**

```python
x = [1, 2, 3]
list(tt.combinations_with_replacement(x, 2))  # (1,1),(1,2),(1,3),(2,2),(2,3),(3,3)

for i in x:
    for j in x:
        if i <= j:
            print((i, j))
```

**sliding window**

```python

def sliding_window(iterable, n):
    it = iter(iterable)
    window = tuple(tt.islice(it, n))
    if len(window) == n:
        yield window
    for x in it:
        window = window[1:] + (x,)
        yield window


list(sliding_window(range(10), 3))
```
