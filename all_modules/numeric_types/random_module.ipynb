**random Module (Python Standard Library)**

**Purpose**

The random module implements pseudo-random number generators for simulations, testing, sampling, shuffling, and probabilistic logic. It is deterministic unless seeded differently.

_**Important**: random is not cryptographically secure. For security-sensitive use cases, use the secrets module._

```python
import random
```

```python

random.seed(10)
random.random()
```

```python
random.random(), random.random()
```

```python
random.randint(1, 100), random.randint(1, 100)
```

```python
random.randrange(1, 100, 20), random.randrange(1, 100, 20)
```

```python
random.uniform(1.5, 5.5), random.uniform(1.5, 5.5)
```

```python
data = [10, 20, 30, 40]

print(f"Original data: {data}")
print(f"Random choice: {random.choice(data)}")
print(f"Random sample of 2: {random.sample(data, 2)}")
print(f"Shuffled data: {random.shuffle(data)}")
print(f"Random Choices: {random.choices(data, k=3)}")
```
