The math module provides fast, C-implemented mathematical functions for real-number arithmetic. It is optimized for performance and precision and is widely used in data engineering, scientific computing, and algorithmic logic.

Key characteristics
* Works with floats (real numbers) only
* Faster than pure Python arithmetic
* Deterministic and precise for mathematical operations
* Does not support complex numbers (use cmath for that)

```python
import math
```

```python
print(f" math.e: {math.e}\n math.pi: {math.pi}\n math.tau: {math.tau} \n math.inf: {math.inf}\n math.nan: {math.nan}")
```

```python
math.pi * 2, math.inf > 10 ** 100
```

```python
print(f"math.sqrt(16): {math.sqrt(16)}\nmath.factorial(5): {math.factorial(5)}\nmath.gcd(48, 18): {math.gcd(48, 18)}")
```

```python
print(f"math.ceil(4.2): {math.ceil(4.2)}\nmath.floor(4.7): {math.floor(4.7)} \nmath.trunc(4.9): {math.trunc(4.9)}")
```

```python
print(math.prod([1, 2, 3, 4, 5]))
```
