**pprint Module (Pretty Printer)**

The pprint (Pretty Print) module in Python is used to print complex Python objects in a readable, well-formatted, and structured manner.

It is particularly useful when dealing with:

* Nested lists and dictionaries

* JSON-like data structures

* Debugging large objects

* Configuration files, API responses, logs

* Data engineering workflows where records are deeply nested

```python
import pprint
```

```python
print(pprint.__all__)
```

```python
data = {"a": 1,
        "b": {"c": 2,
              "d": [10, 20, 30, {"e": 5}]}}
data
```

```python
pprint.pprint(data)
```
