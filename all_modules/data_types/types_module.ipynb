
        types module
        │
        ├── Function & Method Types
        │   ├── FunctionType
        │   ├── LambdaType        --> alias of FunctionType
        │   ├── MethodType        --> bound methods
        │   ├── BuiltinFunctionType (C-implemented functions like len)
        │   └── BuiltinMethodType   (C-implemented methods like dict.get)
        │
        ├── Generator / Coroutine / Async Types
        │   ├── GeneratorType         --> returned by def func(): yield
        │   ├── CoroutineType         --> returned by async def
        │   ├── AsyncGeneratorType    --> async def + yield
        │   ├── FrameType             --> active stack frames
        │   └── TracebackType         --> traceback objects
        │
        ├── Code & Module Types
        │   ├── CodeType          --> compiled bytecode object
        │   ├── ModuleType        --> modules created by import
        │   └── CellType          --> holds closure variables
        │
        ├── Namespace & Descriptor Types
        │   ├── SimpleNamespace       --> lightweight dynamic attribute store
        │   ├── MappingProxyType      --> read-only view of dictionaries
        │   ├── MemberDescriptorType  --> descriptors created via __slots__
        │   └── GetSetDescriptorType  --> built-in descriptors for builtins
        │
        └── Class Construction Utilities
            ├── new_class()          --> low-level class creation
            ├── prepare_class()      --> prepares class namespace
            └── resolve_bases()      --> resolves generic aliases in class bases
