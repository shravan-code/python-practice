**Python enum Module**

The enum module (introduced in Python 3.4) provides support for creating enumerations â€” symbolic names bound to constant, unique values.

Enumerations are used for:
* Readability
* Preventing invalid values
* Creating groups of related constants
* Enforcing immutability and identity checks

```python
import enum
```

```python
class Color(enum.Enum):
    RED = 1
    GREEN = 2
    BLUE = 3
    YELLOW = 4
```

```python
Color.RED.name, Color.RED.value, type(Color)
```

```python
Color.Red = 10
type(Color), type(Color.Red)
```

```python
Color.Red, type(Color), type(Color.Red)
```
